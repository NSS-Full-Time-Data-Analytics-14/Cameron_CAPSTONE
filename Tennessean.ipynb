```python
def give_me_decade_please(Year):
    if 1950 <= Year <= 1959:
        return '1950s'
    elif 1960 <= Year <= 1969:
        return '1960s'
    elif 1970 <= Year <= 1979:
        return '1970s'
    elif 1980 <= Year <= 1989:
        return '1980s'
    elif 1990 <= Year <= 1999:
        return '1990s'
    elif 2000 <= Year <= 2009:
        return '2000s'
    elif 2010 <= Year <= 2019:
        return '2010s'
    elif 2020 <= Year <= 2029:
        return '2020s'
    else:
        return 'Unknown'

tn_torn['Decade'] = tn_torn['Year'].apply(give_me_decade_please)
tn_torn
```

```python
fatalities_per_decade = tn_torn.groupby('Decade')['Fatality'].sum().reset_index()
fatalities_per_decade
```

```python
injuries_per_decade = tn_torn.groupby('Decade')['Injury'].sum().reset_index()
injuries_per_decade
```
