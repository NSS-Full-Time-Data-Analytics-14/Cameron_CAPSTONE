```python
def give_me_decade_please(Year):
    if 1950 <= Year <= 1959:
        return '1950s'
    elif 1960 <= Year <= 1969:
        return '1960s'
    elif 1970 <= Year <= 1979:
        return '1970s'
    elif 1980 <= Year <= 1989:
        return '1980s'
    elif 1990 <= Year <= 1999:
        return '1990s'
    elif 2000 <= Year <= 2009:
        return '2000s'
    elif 2010 <= Year <= 2019:
        return '2010s'
    elif 2020 <= Year <= 2029:
        return '2020s'
    else:
        return 'Unknown'

tn_torn['Decade'] = tn_torn['Year'].apply(give_me_decade_please)
tn_torn
```

```python
def get_month_name(month):
    if month == 1:
        return "January"
    elif month == 2:
        return "February"
    elif month == 3:
        return "March"
    elif month == 4:
        return "April"
    elif month == 5:
        return "May"
    elif month == 6:
        return "June"
    elif month == 7:
        return "July"
    elif month == 8:
        return "August"
    elif month == 9:
        return "September"
    elif month == 10:
        return "October"
    elif month == 11:
        return "November"
    elif month == 12:
        return "December"
    else:
        return "something_else"

tn_torn['month_name'] = tn_torn['month'].apply(get_month_name)

tn_torn[['month', 'month_name']].head()
```

```python
from sqlalchemy import create_engine, text
```

```python
tornado_sql_database = 'Tornado'    

tornado_connection_string = f"postgresql://postgres:postgres@localhost:5432/{tornado_sql_database}"
```

```python
engine = create_engine(tornado_connection_string)
```

```python
query_1 = '''SELECT ef_scale, ROUND(AVG(width_mi), 2) AS recorded_avg_size, ROUND(MAX(width_mi), 2) AS max_size, 
		ROUND(MIN(width_mi), 2) AS min_size
FROM tornado
GROUP BY ef_scale
ORDER BY recorded_avg_size DESC'''
```

```python
with engine.connect() as connection:
    average_sizes = pd.read_sql(text(query_1), con = connection) 

average_sizes.head()
```

```python
query_2 = '''SELECT ef_scale, COUNT(*), ROUND(AVG(path_length_mi),2) AS avg_path, ROUND(AVG(width_mi),2) AS avg_width_ft, 
		ROUND(AVG(property_dmg),2) AS avg_damage_cost, ROUND(AVG(injury)) AS avg_ouchies, ROUND(AVG(fatality)) AS avg_deaths,
  SUM(fatality) as total_deaths
FROM tornado
Group BY ef_scale
ORDER BY avg_width_ft DESC'''
```

```python
with engine.connect() as connection:
    all_averages = pd.read_sql(text(query_2), con = connection) 

all_averages.head()
```

```python
all_averages.dtypes
```

```python
max_size_query = '''SELECT date, SUM(injury) AS injuries, MAX(width_mi) AS max_size_ft
FROM tornado
GROUP BY date'''
```

```python
with engine.connect() as connection:
    max_size_per_date = pd.read_sql(text(max_size_query), con = connection) 

max_size_per_date.head()
```
