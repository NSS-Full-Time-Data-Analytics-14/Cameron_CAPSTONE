```python
def give_me_decade_please(Year):
    if 1950 <= Year <= 1959:
        return '1950s'
    elif 1960 <= Year <= 1969:
        return '1960s'
    elif 1970 <= Year <= 1979:
        return '1970s'
    elif 1980 <= Year <= 1989:
        return '1980s'
    elif 1990 <= Year <= 1999:
        return '1990s'
    elif 2000 <= Year <= 2009:
        return '2000s'
    elif 2010 <= Year <= 2019:
        return '2010s'
    elif 2020 <= Year <= 2029:
        return '2020s'
    else:
        return 'Unknown'

tn_torn['Decade'] = tn_torn['Year'].apply(give_me_decade_please)
tn_torn
```

```python
torns_yearly = tn_torn.groupby('Year').size().reset_index(name = 'tornado_reports')
type(torns_yearly)
```

```python
torns_yearly = torns_yearly.drop(0).reset_index()
torns_yearly
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(x="Year", y="tornado_reports", data = torns_yearly, palette="plasma")

plt.title("Tornado Accounts by Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Tornado Count", fontsize=12)
plt.xticks(rotation = 95)  
plt.tight_layout();
```

```python
Injuries_per_year = tn_torn.groupby('Year')['Injury'].sum().reset_index()
Injuries_per_year = Injuries_per_year.drop(0).reset_index()
Injuries_per_year = Injuries_per_year.reset_index(drop = True)
Injuries_per_year = Injuries_per_year.drop(columns=['index'])
Injuries_per_year
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(x="Year", y="Injury", data = Injuries_per_year, palette="plasma")

plt.title("Tornado Accounts by Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Injuries Caused", fontsize=12)
plt.xticks(rotation = 95)  
plt.tight_layout();
```

```python
fatalities_per_year = tn_torn.groupby('Year')['Fatality'].sum().reset_index()
fatalities_per_year.sort_values(by = 'Fatality', ascending = False).head()
fatalities_per_year = fatalities_per_year.drop(0)
fatalities_per_year
```

```python
fatalities_per_decade = tn_torn.groupby('Decade')['Fatality'].sum().reset_index()
fatalities_per_decade = fatalities_per_decade.drop(8)
fatalities_per_decade
```

```python
sns.scatterplot(data=fatalities_per_decade, x="Decade", y="Fatality", color = 'black')
```

```python
injuries_per_decade = tn_torn.groupby('Decade')['Injury'].sum().reset_index()
injuries_per_decade = injuries_per_decade.drop(8)
injuries_per_decade
```

```python
sns.scatterplot(data=injuries_per_decade, x="Decade", y="Injury", color = 'green')
```

```python
merged = pd.merge(injuries_per_decade, fatalities_per_decade, on = 'Decade')
merged
```

```python
merged.describe()
```

```python
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=merged, kind="bar",
    x="Injury", y="Fatality", hue="Decade",
    errorbar="sd", palette="dark", alpha=.6, height=6)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```
